```python
import os
import pandas as pd
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits import mplot3d
from matplotlib.ticker import MaxNLocator
from AnalyserTools import AnalyserTools as ant
```

```python
path_dir_0: str = r"/home/svergani/monitoring/files/rmsm_display_0/"
path_dir_1: str = r"/home/svergani/monitoring/files/rmsm_display_1/"
path_dir_2: str = r"/home/svergani/monitoring/files/rmsm_display_2/"
```

```python
#I start by opening one file and check how many channels are stored there
f_0 = h5.File(os.sep.join([path_dir_0, "rmsm_display-0-220525-151538.hdf5"]), "r")
channels_0 = np.array(f_0.get('data/axis0').value)
f_1 = h5.File(os.sep.join([path_dir_1, "rmsm_display-1-220525-151538.hdf5"]), "r")
channels_1 = np.array(f_1.get('data/axis0').value)
f_2 = h5.File(os.sep.join([path_dir_2, "rmsm_display-2-220525-151538.hdf5"]), "r")
channels_2 = np.array(f_2.get('data/axis0').value)
```

```python
#Set default figure size
mpl.rcParams['figure.figsize'] = [30, 20]
mpl.rcParams["legend.frameon"] = False
mpl.rcParams.update({'font.size': 22})
```

```python
time = np.arange(std_0.shape[0])*10#measurements are saved with a delay of 11 s but technically is 10 s
#Plot the first waveforms
fig, ax = plt.subplots()
ax.plot(time,std_0[:,3],label="Input Pulse")
ax.set_xlabel("Time [s]")
ax.set_ylabel("STD")
ax.set_title("Distribution of STD for Channel 3 over time on Display 0")
```

```python
#Plot all channels together
fig, ax = plt.subplots()
for i in range(std_0.shape[1]):
    ax.plot(time,std_0[:,i],label=i)
ax.set_xlabel("Time [s]")
ax.set_ylabel("STD")
ax.set_title("Distribution of STD for All Channels over time in Display 0")
```

```python
time_0 = np.arange(std_0.shape[0])*10
time_1 = np.arange(std_1.shape[0])*10
time_2 = np.arange(std_2.shape[0])*10
```

```python
#this is my way to create the channels to be plotted on the 3d plot
#this will loose the information about the channel number
#but no hole

channels_3d_0 = []
for i in range(channels_0.size):
    channels_3d_0.append(np.ones(time_0.size,dtype=int)*i)
channels_3d_0 = np.array(channels_3d_0)

channels_3d_1 = []
for i in range(channels_1.size):
    channels_3d_1.append(np.ones(time_1.size,dtype=int)*i)
channels_3d_1 = np.array(channels_3d_1)

channels_3d_2 = []
for i in range(channels_2.size):
    channels_3d_2.append(np.ones(time_2.size,dtype=int)*i)
channels_3d_2 = np.array(channels_3d_2)
```

```python
###I want the colour scale to be related to the channel with the highest value.
#I loop over all channel looking for the highest value of std

fig, ax = plt.subplots()
highest_value = 0
for i in range(std_0.shape[1]):
    if(np.amax(std_0[:,i])>highest_value):
        highest_value=np.amax(std_0[:,i])
for i in range(std_0.shape[1]):
    sc=ax.scatter(time_0, channels_3d_0[i,:], c=std_0[:,i],vmin=0, vmax=highest_value,marker=',',s=140)
    
normalize = mpl.colors.Normalize(vmin=0, vmax=highest_value)                            
cbar = fig.colorbar(sc)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Channel Number")
ax.yaxis.set_major_locator(MaxNLocator(integer=True))#forces it to take only integer values
ax.set_title("Distribution of STD for All Channels over time in Display 0")
plt.show()
```

```python
fig, ax = plt.subplots()
highest_value = 0
for i in range(std_1.shape[1]):
    if(np.amax(std_1[:,i])>highest_value):
        highest_value=np.amax(std_1[:,i])
for i in range(std_1.shape[1]):
    sc=ax.scatter(time_1, channels_3d_1[i,:], c=std_1[:,i],vmin=0, vmax=highest_value,marker=',',s=140)
    
normalize = mpl.colors.Normalize(vmin=0, vmax=highest_value)                            
cbar = fig.colorbar(sc)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Channel Number")
ax.yaxis.set_major_locator(MaxNLocator(integer=True))#forces it to take only integer values
ax.set_title("Distribution of STD for All Channels over time in Display 1")
plt.show()
```

```python
fig, ax = plt.subplots()
highest_value = 0
for i in range(std_2.shape[1]):
    if(np.amax(std_2[:,i])>highest_value):
        highest_value=np.amax(std_2[:,i])
for i in range(std_2.shape[1]):
    sc=ax.scatter(time_2, channels_3d_2[i,:], c=std_2[:,i],vmin=0, vmax=highest_value,marker=',',s=140)
    
normalize = mpl.colors.Normalize(vmin=0, vmax=highest_value)                            
cbar = fig.colorbar(sc)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Channel Number")
ax.yaxis.set_major_locator(MaxNLocator(integer=True))#forces it to take only integer values
ax.set_title("Distribution of STD for All Channels over time in Display 2")
plt.show()
```
