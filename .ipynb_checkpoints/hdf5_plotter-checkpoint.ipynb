```python
####how to open an HDF5 file, extract data, and start some basics data analysis


if __name__ == '__main__':
# import required libraries
    import h5py as h5
    import numpy as np
    import matplotlib.pyplot as plt
    
    # Read H5 file
    f = h5.File("/home/svergani/monitoring/test_results/alarms_2022_10_28_15_50_19.hdf5", "r")
    # Get and print list of datasets within the H5 file
    datasetNames = [n for n in f.keys()]
    for n in datasetNames:
        print(n)
```

```python
alarm_pixel_0 = np.array(f.get('alarm_pixel_0').value)
alarm_pixel_1 = np.array(f.get('alarm_pixel_1').value)
alarm_pixel_2 = np.array(f.get('alarm_pixel_2').value)
channels_0 = np.array(f.get('channels_0').value)
channels_1 = np.array(f.get('channels_1').value)
channels_2 = np.array(f.get('channels_2').value)
```

```python
fig, ax = plt.subplots()
ax.plot(channels_0,alarm_pixel_0,label="Input Pulse")
ax.set_xlabel("Channel number")
ax.set_ylabel("Trigger Value")
ax.set_title("Triggers for view 0")
```

```python
fig, ax = plt.subplots()
ax.plot(channels_1,alarm_pixel_1,label="Input Pulse")
ax.set_xlabel("Channel number")
ax.set_ylabel("Trigger Value")
ax.set_title("Triggers for view 1")
```

```python
fig, ax = plt.subplots()
ax.plot(channels_2,alarm_pixel_2,label="Input Pulse")
ax.set_xlabel("Channel number")
ax.set_ylabel("Trigger Value")
ax.set_title("Triggers for view 2")
```
