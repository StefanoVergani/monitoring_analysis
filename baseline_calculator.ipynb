```python
#first thing we want to call some libraries which will be useful 

import os
import pandas as pd
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits import mplot3d
from matplotlib.ticker import MaxNLocator
```

```python
#here is the path to where your baseline files are
#remember: at the moment we need one baseline file for each display

path_dir_baseline: str = r"/home/svergani/monitoring/hdf5_created_files"
bl_0 = h5.File(os.sep.join([path_dir_baseline, "rmsm_display_0_2022_05_22.hdf5"]), 'r')
bl_1 = h5.File(os.sep.join([path_dir_baseline, "rmsm_display_1_2022_05_22.hdf5"]), 'r')
bl_2 = h5.File(os.sep.join([path_dir_baseline, "rmsm_display_2_2022_05_22.hdf5"]), 'r')
```

```python
#extracting the array

rmsm_bl_0 = np.array(bl_0.get('rmsm').value)
rmsm_bl_1 = np.array(bl_1.get('rmsm').value)
rmsm_bl_2 = np.array(bl_2.get('rmsm').value)
channels_bl_0 = np.array(bl_0.get('channels').value)
channels_bl_1 = np.array(bl_1.get('channels').value)
channels_bl_2 = np.array(bl_2.get('channels').value)
```

```python
time_0 = np.arange(rmsm_bl_0.shape[0])*10
time_1 = np.arange(rmsm_bl_1.shape[0])*10
time_2 = np.arange(rmsm_bl_2.shape[0])*10
```

```python
#this is my way to create the channels to be plotted on the 3d plot
#weird but it works

channels_3d_0 = []
for i in range(np.amin(channels_bl_0),np.amax(channels_bl_1)):
    channels_3d_0.append(np.ones(time_0.size,dtype=int)*i)
channels_3d_0 = np.array(channels_3d_0)

channels_3d_1 = []
for i in range(np.amin(channels_bl_1),np.amax(channels_bl_1)):
    channels_3d_1.append(np.ones(time_1.size,dtype=int)*i)
channels_3d_1 = np.array(channels_3d_1)

channels_3d_2 = []
for i in range(np.amin(channels_bl_2),np.amax(channels_bl_2)):
    channels_3d_2.append(np.ones(time_2.size,dtype=int)*i)
channels_3d_2 = np.array(channels_3d_2)
```

```python
#now it is time to plot what we have to check that we are actually doing things right

#Set default figure size
mpl.rcParams['figure.figsize'] = [30, 20]
mpl.rcParams["legend.frameon"] = False
mpl.rcParams.update({'font.size': 22})
```

```python
###I want the colour scale to be related to the channel with the highest value.
#I loop over all channel looking for the highest value of rmsm

fig, ax = plt.subplots()
highest_value = 0
for i in range(rmsm_bl_0.shape[1]):
    if(np.amax(rmsm_bl_0[:,i])>highest_value):
        highest_value=np.amax(rmsm_bl_0[:,i])
for i in range(rmsm_bl_0.shape[1]):
    sc=ax.scatter(time_0, channels_3d_0[i,:], c=rmsm_bl_0[:,i],vmin=0, vmax=highest_value,marker=',',s=140)
    
normalize = mpl.colors.Normalize(vmin=0, vmax=highest_value)                            
cbar = fig.colorbar(sc)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Channel Number")
ax.yaxis.set_major_locator(MaxNLocator(integer=True))#forces it to take only integer values
ax.set_title("Distribution of RMSM for All Channels over time in Display 0")
plt.show()
```

```python
fig, ax = plt.subplots()
highest_value = 0
for i in range(rmsm_bl_1.shape[1]):
    if(np.amax(rmsm_bl_1[:,i])>highest_value):
        highest_value=np.amax(rmsm_bl_1[:,i])
for i in range(rmsm_bl_1.shape[1]):
    sc=ax.scatter(time_1, channels_3d_1[i,:], c=rmsm_bl_1[:,i],vmin=0, vmax=highest_value,marker=',',s=140)
    
normalize = mpl.colors.Normalize(vmin=0, vmax=highest_value)                            
cbar = fig.colorbar(sc)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Channel Number")
ax.yaxis.set_major_locator(MaxNLocator(integer=True))#forces it to take only integer values
ax.set_title("Distribution of RMSM for All Channels over time in Display 1")
plt.show()
```

```python
fig, ax = plt.subplots()
highest_value = 0
for i in range(rmsm_bl_2.shape[1]):
    if(np.amax(rmsm_bl_2[:,i])>highest_value):
        highest_value=np.amax(rmsm_bl_2[:,i])
for i in range(rmsm_bl_2.shape[1]):
    sc=ax.scatter(time_2, channels_3d_2[i,:], c=rmsm_bl_2[:,i],vmin=0, vmax=highest_value,marker=',',s=140)
    
normalize = mpl.colors.Normalize(vmin=0, vmax=highest_value)                            
cbar = fig.colorbar(sc)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Channel Number")
ax.yaxis.set_major_locator(MaxNLocator(integer=True))#forces it to take only integer values
ax.set_title("Distribution of RMSM for All Channels over time in Display 2")
plt.show()
```

```python
#now we evaluate the baseline for each channel and the baseline for each display
#at this stage, it is just a lazy mean but later on I can do something more fancy is required

mean_display_0 = []
mean_display_1 = []
mean_display_2 = []

for i in range(rmsm_bl_0.shape[1]):
    mean_display_0.append(np.mean(rmsm_bl_0[:,i]))
    
for i in range(rmsm_bl_1.shape[1]):
    mean_display_1.append(np.mean(rmsm_bl_1[:,i]))
    
for i in range(rmsm_bl_2.shape[1]):
    mean_display_2.append(np.mean(rmsm_bl_2[:,i]))
    
mean_display_0 = np.array(mean_display_0)
mean_display_1 = np.array(mean_display_1)
mean_display_2 = np.array(mean_display_2)
#now let's do a mean of all the mean arrays inside a display
#that is basically the average value of the entire display
#it is not the smartest way and it will be changed, but for now I use this

total_mean_0 = np.mean(mean_display_0)
total_mean_1 = np.mean(mean_display_1)
total_mean_2 = np.mean(mean_display_2)
```

```python
#let's plot the means just to check what we are doing
#you might wonder why I am creating the arrays channels0 etc. when we already have our 
#arrays with the actual channel number. the problem is than the curves are in different
#points and it becomes hard to visualise them. For this reason, let them start all from zero.
channels0 = np.arange(mean_display_0.shape[0])
channels1 = np.arange(mean_display_1.shape[0])
channels2 = np.arange(mean_display_2.shape[0])


fig, ax = plt.subplots()
ax.plot(channels0,mean_display_0, label="Mean Values for Channels in Display 0")
ax.plot(channels1,mean_display_1, label="Mean Values for Channels in Display 1")
ax.plot(channels2,mean_display_2, label="Mean Values for Channels in Display 2")
ax.scatter(channels0.size/2, total_mean_0, marker="s",s=200, label="Mean for Display 0")
ax.scatter(channels1.size/2, total_mean_1, marker="s",s=200, label="Mean for Display 1")
ax.scatter(channels2.size/2, total_mean_2, marker="s",s=200, label="Mean for Display 2")
#for i in range(mean_display_0.shape[0]):
#    ax.plot(channels[i],mean_display_0[i],label=i)
ax.set_xlabel("Channels")
ax.set_ylabel("Mean ADC")
ax.set_title("Mean ADC for All Channels in Dispaly 0")
ax.legend(fontsize=30,loc='upper right')
```

```python
#now I store the baseline values as hdf5 files

hf_target = h5.File(os.sep.join([path_dir_baseline, "baselines_2022_05_22.hdf5"]), 'w')
```

```python
hf_target.create_dataset('mean_display_0', data=mean_display_0)
hf_target.create_dataset('mean_display_1', data=mean_display_1)
hf_target.create_dataset('mean_display_2', data=mean_display_2)
hf_target.create_dataset('total_mean_0', data=total_mean_0)
hf_target.create_dataset('total_mean_1', data=total_mean_1)
hf_target.create_dataset('total_mean_2', data=total_mean_2)
hf_target.create_dataset('channels_bl_0', data=channels_bl_0)
hf_target.create_dataset('channels_bl_1', data=channels_bl_1)
hf_target.create_dataset('channels_bl_2', data=channels_bl_2)
```
